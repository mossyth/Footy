```python
#example of how to get event/tracking data using statsbombpy API
from statsbombpy import sb
```

```python
#list all the competitions in the database
df = sb.competitions()

#only show 15/16 comps
df = df[(df["season_name"] == "2015/2016")]
df
```

```python
#all matches in the premier league 15/16 season
matches = sb.matches(competition_id=2, season_id=27)

#all of leicesters away games that season
leicesterAwayMatches = matches[(matches['away_team'] == 'Leicester City')]
leicesterAwayMatches
```

```python
#events for the man city v leicester city 15/16 game
events = sb.events(match_id=3754290)
events.head()
```

```python
events.columns
```

```python
#only use the column types we need 
events = events[['team' , 'type', 'location', 'pass_end_location', 'player']]
events = events[events['team'] == 'Leicester City'].reset_index()

events
```
