```python
import pandas as pd
import requests #The requests module allows you to send HTTP requests using Python.
import time
```

```python
standings_url = 'https://fbref.com/en/comps/9/Premier-League-Stats'
```

```python
data = requests.get(standings_url)
```

```python
data.text
```

```python
%pip install BeautifulSoup4 selenium
from bs4 import BeautifulSoup

#allows us to parse the html. 
#It creates a parse tree for documents that can be used to extract data from HTML, which is useful for web scraping.
```

```python
soup = BeautifulSoup(data.text)
standings_table = soup.select('table.stats_table')[0] #css selector to select table with class name stats_table

links = [l.get('href') for l in standings_table.find_all('a')]
links = [l for l in links if '/squads/' in l]  #only want the links with /squads/ in it
team_urls = [f"https://fbref.com{l}" for l in links] #create list of all the teams urls/links
team_urls
```

```python
all_players = []
```

```python
data = requests.get(team_urls[0])
print(data.text)        
```

```python
import time
for team_url in team_urls:
        team_name = team_url.split("/")[-1].replace("-Stats", "").replace("-", "")

        data = requests.get(team_url)
        standings_table = soup.select('table.stats_table')[0] #css selector to select table with class name stats_table
        defensivePlayerActions = pd.read_html(data.text, match="Defensive Actions")[0]

        all_players.append(defensivePlayerActions)
        time.sleep(1)
```

```python
all_players_df = pd.concat(all_players)
all_players_df
```

```python
all_players_df.columns.values
```

```python
all_players_df.columns = ['_'.join(col).strip() for col in all_players_df.columns.values]
all_players_df.columns
```

```python
all_players_df = all_players_df.drop(["Challenges_Tkl", "Challenges_Att", "Challenges_Tkl%", "Challenges_Lost", "Unnamed: 21_level_0_Matches"], axis=1)
```

```python
all_players_df
```

```python
data_with_index = all_players_df.set_index("Unnamed: 0_level_0_Player")
data_with_index = data_with_index.drop(["Squad Total", "Opponent Total"])

data_with_index = data_with_index.reset_index()
df = data_with_index
df
```

```python
df.to_csv("defensiveActions.csv")
```
