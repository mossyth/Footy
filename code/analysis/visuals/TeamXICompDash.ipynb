```python
ATTdf = ATTdf.drop('Unnamed: 0', axis=1)
MIDdf = MIDdf.drop('Unnamed: 0', axis=1)
DEFdf = DEFdf.drop('Unnamed: 0', axis=1)
GKdf = GKdf.drop('Unnamed: 0', axis=1)
```

```python
def plotRadarComp(df, playerA, playerB):
    params = list(df.columns)
    params = params[2:-1]

    ranges = []
    a_values = []
    b_values = []

    for x in params:
        y = [i for i in df[params][x] if i != 0]
        
        a = min(y)
        a -= (a*.15)

        b = max(y)
        b += (b*.15)

        ranges.append((a,b))
    
    for x in range(len(df['Player'])):
        if (df['Player'][x] == playerA):
            a_values = df.iloc[x].values.tolist()
        if (df['Player'][x] == playerB):
            b_values = df.iloc[x].values.tolist()

    a_values = a_values[2:-1]
    b_values = b_values[2:-1]
    values = [a_values, b_values]

    title = dict(
    title_name = playerA,
    title_color = 'blue',
    title_name_2 = playerB,
    title_color_2 = 'red',
    title_fontsize = 18)
    endnote = '\n mossystats'

    filename=f"{playerA.split()[-1]}vs{playerB.split()[-1]}.jpg"
    
    radar = Radar(background_color="black", patch_color="#28252C", label_color="#F0FFF0",
                range_color="#F0FFF0")

    fig, ax = radar.plot_radar(ranges=ranges, params=params, values=values, 
                            radar_color = ['blue','red'], alphas=[0.8, 0.6], title=title, 
                            endnote=endnote, end_color='green', dpi=700, alpha= 0.7,
                            image='C:/Users/tomas.hehir/Source/Repos/Footy/images/PLLogo.png', image_coord=[0.495, 0.805, 0.04, 0.1], 
                            filename=f"C:/Users/tomas.hehir/Source/Repos/Footy/outputs/PL/{filename}",
                            compare=True)
```

```python
plotRadarComp(GKdf, "Emiliano Martnez", "Guglielmo Vicario")

plotRadarComp(DEFdf, "Kyle Walker", "Ben White")
plotRadarComp(DEFdf, "Virgil van Dijk", "William Saliba")
plotRadarComp(DEFdf, "Nathan Ak", "Gabriel Magalhes")
plotRadarComp(DEFdf, "Joko Gvardiol", "Destiny Udogie")

plotRadarComp(MIDdf, "Alexis Mac Allister", "Douglas Luiz")
plotRadarComp(MIDdf, "Rodri", "Declan Rice")
plotRadarComp(MIDdf, "Kevin De Bruyne", "Martin degaard")

plotRadarComp(ATTdf, "Phil Foden", "Bukayo Saka")
plotRadarComp(ATTdf, "Darwin Nez", "Kai Havertz")
plotRadarComp(ATTdf, "Erling Haaland", "Ollie Watkins")
plotRadarComp(ATTdf, "Luis Daz", "Leandro Trossard")
```
