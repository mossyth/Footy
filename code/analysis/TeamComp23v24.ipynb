```python
def plotComp(df, playerA, seasonA, playerB, seasonB):
    params = list(df.columns)
    params = params[5:]

    ranges = []
    a_values = []
    b_values = []

    for x in params:
        a = min(df[params][x])
        a -= (a*.15)

        b = max(df[params][x])
        b += (b*.15)

        ranges.append((a,b))

    for x in range(len(df['Player'])):
        if (df['Player'][x] == playerA) & (df['Season'][x] == seasonA):
            a_values = df.iloc[x].values.tolist()
        if (df['Player'][x] == playerB) & (df['Season'][x] == seasonB):
            b_values = df.iloc[x].values.tolist()

    a_values = a_values[5:]
    b_values = b_values[5:]
    values = [a_values, b_values]

    title = dict(
    title_name = playerA,
    title_color = 'white',
    subtitle_name='2022/23',
    subtitle_color='white',
    title_name_2 = playerB,
    title_color_2 = 'red',
    subtitle_name_2='2023/24',
    subtitle_color_2='red',
    title_fontsize = 18)
    endnote = '\n mossystats'

    filename=f"{playerA.split()[-1]}{seasonA}vs{playerB.split()[-1]}{seasonB}.jpg"
    
    radar = Radar(background_color="black", patch_color="#28252C", label_color="#F0FFF0",
                range_color="#F0FFF0")

    fig, ax = radar.plot_radar(ranges=ranges, params=params, values=values, 
                            radar_color = ['white','red'], alphas=[0.8, 0.6], title=title, 
                            endnote=endnote, end_color='green', dpi=700, alpha= 0.7,
                            image='C:/Users/tomas.hehir/Source/Repos/Footy/images/arsenalcrest.jpg', image_coord=[0.495, 0.805, 0.04, 0.1], 
                            filename=f"C:/Users/tomas.hehir/Source/Repos/Footy/outputs/arsenal/{filename}",
                            compare=True)
```

```python
plotComp(GKdf, "Aaron Ramsdale", 2023, "David Raya", 2024)

plotComp(DEFdf, "Ben White", 2023, "Ben White", 2024)
plotComp(DEFdf, "William Saliba", 2023, "William Saliba", 2024)
plotComp(DEFdf, "Gabriel Magalhães", 2023, "Gabriel Magalhães", 2024)
plotComp(DEFdf, "Oleksandr Zinchenko", 2023, "Oleksandr Zinchenko", 2024)

plotComp(MIDdf, "Thomas Partey", 2023, "Jorginho", 2024)
plotComp(MIDdf, "Granit Xhaka", 2023, "Declan Rice", 2024)
plotComp(MIDdf, "Martin Ødegaard", 2023, "Martin Ødegaard", 2024)

plotComp(ATTdf, "Bukayo Saka", 2023, "Bukayo Saka", 2024)
plotComp(ATTdf, "Gabriel Jesus", 2023, "Kai Havertz", 2024)
plotComp(ATTdf, "Gabriel Martinelli", 2023, "Leandro Trossard", 2024)
```
