```python
df = df[df['90s'] >  10.0]
df = df.reset_index()
df = df.drop("index", axis=1)
df
```

```python
#get parameters
params = list(df.columns)
params = params[5:]
params
```

```python
ranges = []
a_values = []
b_values = []


for x in params:
    print(x)
    a = min(df[params][x])
    print(a)

    b = max(df[params][x])
    print(b)

    ranges.append((a,b))
```

```python
df = df[(df['Player'] == "Sergio Ramos") | (df['Player'] == "Antonio Rüdiger")]
df
```

```python
df = df.reset_index()
df
```

```python
df = df.drop('index', axis=1)
```

```python
#values
for x in range(len(df['Player'])):
    if df['Player'][x] == 'Antonio Rüdiger':
        a_values = df.iloc[x].values.tolist()
    if df['Player'][x] == 'Sergio Ramos':
        b_values = df.iloc[x].values.tolist()

a_values = a_values[5:]
b_values = b_values[5:]
values = [a_values, b_values]
```
