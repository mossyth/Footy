```python
import pandas as pd
import matplotlib.pyplot as plt
from soccerplots.radar_chart import Radar
from soccerplots.utils import add_image
```

```python
df = pd.read_csv('C:/Users/tomas.hehir/Source/Repos/Footy/data/PLDefenderP90.csv')
```

```python
df.head()
```

```python
df = df.drop('Unnamed: 0', axis=1)
```

```python
df
```

```python
df.columns = ['Player', 'Nation', 'Position', 'Age', '90s', 'Tackles', 'TacklesWon', 'TacklesDef3rd',
'TacklesMid3rd', 'TacklesAtt3rd', 'Blocks', 'ShotsBlocked', 'PassesBlocked', 'Intercptions', 'Tkl+Int', 'Clearances', 'Errors']
```

```python
df
```

```python
df = df[(df['Player'] == "Gabriel Magalhães") | (df['Player'] == "Cristian Romero")]
df
```

```python
df = df.reset_index()
df
```

```python
df = df.drop('index', axis=1)
```

```python
df
```

```python
#get parameters
params = list(df.columns)
params = params[5:]
params
```

```python
ranges = []
a_values = []
b_values = []


for x in params:
    print(x)
    a = min(df[params][x])
    print(a)
    a = a - (a*0.25)

    b = max(df[params][x])
    b = b + (b*0.25)

    ranges.append((a,b))

#values
for x in range(len(df['Player'])):
    if df['Player'][x] == 'Gabriel Magalhães':
        a_values = df.iloc[x].values.tolist()
    if df['Player'][x] == 'Cristian Romero':
        b_values = df.iloc[x].values.tolist()

a_values = a_values[5:]
b_values = b_values[5:]
values = [a_values, b_values]
```

```python
values
```

```python
#title 

title = dict(
    title_name = 'Gabriel Magalhães',
    title_color = '#B6282F',
    subtitle_name='Arsenal',
    subtitle_color='#B6282F',
    title_name_2 = 'Cristian Romero',
    title_color_2 = '#344D94',
    subtitle_name_2='Spurs',
    subtitle_color_2='#344D94',
    title_fontsize = 18)
endnote = '\n mossystats'
```

```python
radar = Radar(background_color="black", patch_color="#28252C", label_color="#F0FFF0",
              range_color="#F0FFF0")

fig, ax = radar.plot_radar(ranges=ranges, params=params, values=values, 
                          radar_color = ['#B6282F','#344D94'],
                           alphas=[0.8, 0.6], title=title, endnote=endnote, end_color='green', 
                           dpi=700, alpha= 0.7, compare=True)

fig = add_image(image="C:/Users/tomas.hehir/Source/Repos/Footy/images/gabriel.png", fig=fig, left=0.305, bottom=0.74, width=0.1, height=0.075)
fig = add_image(image="C:/Users/tomas.hehir/Source/Repos/Footy/images/romero.png", fig=fig, left=0.62, bottom=0.74, width=0.1, height=0.075)
```
