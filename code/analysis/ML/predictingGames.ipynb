```python
matches["date"] = pd.to_datetime(matches["date"])
```

```python
matches["venue_code"] = matches["venue"].astype("category").cat.codes #convert from string to categories, then categories into numbers
```

```python
matches["opp_code"] = matches["opponent"].astype("category").cat.codes
```

```python
matches["hour"] = matches["time"].str.replace(":.+", "", regex=True).astype("int")
```

```python
matches["day_code"] = matches["date"].dt.dayofweek
```

```python
matches["target"] = (matches["result"] == "W").astype("int")
```

```python
matches.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
#A random forest is a meta estimator that fits a number of decision tree classifiers on various sub-samples of the dataset
#and uses averaging to improve the predictive accuracy and control over-fitting. Trees in the forest use the best split strategy
#i.e. equivalent to passing splitter="best" to the underlying DecisionTreeRegressor.
#The sub-sample size is controlled with the max_samples parameter if bootstrap=True (default), 
#otherwise the whole dataset is used to build each tree.
#n_estimators - The number of trees in the forest.


rf = RandomForestClassifier(n_estimators=50, min_samples_split=10, random_state=1)
```

```python
train = matches[matches["date"] < '2024-05-18']
```

```python
test = matches[matches["date"] > '2024-05-18']
```

```python
predictors = ["venue_code", "opp_code", "hour", "day_code"]
#predictors = ["venue_code", "opp_code"]
```

```python
rf.fit(train[predictors], train["target"])
```

```python
preds= rf.predict(test[predictors])
```

```python
from sklearn.metrics import accuracy_score 
```

```python
acc = accuracy_score(test["target"], preds)
```

```python
#create more predictors, to improve accuracy of our model. Lets try and look at teams current form

#lets create a df for every squad in our data
grouped_matches = matches.groupby("team")
```

```python
group = grouped_matches.get_group("Arsenal")
group.columns
```

```python
def rolling_averages(group, cols, new_cols):
    group= group.sort_values("date")
    rolling_stats = group[cols].rolling(5, closed="left").mean()
    group[new_cols] = rolling_stats
    group = group.dropna(subset=new_cols)
    return group
```

```python
cols = ["gf", "ga", "sh", "sot", "dist", "xg", "xga" ]
new_cols = [f"{c}_rolling" for c in cols]
```

```python
new_cols #new columsn we want to create with the rolling avgs
```

```python
rolling_averages(group, cols, new_cols) # for Arsenal
```

```python
matches_rolling = matches.groupby("team").apply(lambda x: rolling_averages(x, cols, new_cols))
```

```python
matches_rolling
```

```python
matches_rolling = matches_rolling.droplevel('team')
```

```python
matches_rolling.columns
```

```python
matches_rolling.index = range(matches_rolling.shape[0]) #reset the indices
```

```python
matches_rolling
```

```python
#predictors = ["venue_code", "opp_code", "hour", "day_code", "gf_rolling"]
new_cols
```

```python
def make_predictions(data, predictors):
    train = data[data["date"] < '2024-05-18']
    test = data[data["date"] > '2024-05-18']
    rf.fit(train[predictors], train["target"])
    preds= rf.predict(test[predictors])
    combined = pd.DataFrame(dict(actual = test["target"], predicted = preds), index=test.index)
    precision = precision_score(test["target"], preds)
    acc = accuracy_score(test["target"], preds)
    return combined, precision, acc
```

```python
combined, precision, acc = make_predictions(matches_rolling, predictors + new_cols)
```
